```python
import json
from pathlib import Path
import numpy as np
```

```python
data: dict = {}
Ntotal = 61
for i in range(Ntotal):
    d = json.loads(Path(f'./transforms_{i:02d}.json').read_text())
    for frame in d['frames']:
        frame['time'] = i/(Ntotal-1)
        
    if i == 0:
        data.update(d)
        continue
    elif i==Ntotal-1:
        data['frames'].extend(d['frames'])
        continue
    
    for frame in d['frames']:
        if 'eval' in frame['file_path']:
            data['frames'].append(frame)
            continue
        p = Path(frame['file_path'])
        ind = int(p.stem.split('_')[-1])
        if i%10 == 0:
            inds = [1,33]
        else:
            inds = [1,4]
        if ind in inds:
            data['frames'].append(frame)

Path('./transforms_0_to_60.json').write_text(json.dumps(data, indent=4))
```
